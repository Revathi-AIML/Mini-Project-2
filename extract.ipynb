```python
import os
import yaml
import pandas as pd
from pathlib import Path
data_dir = r"F:\vscode\stock_analysis\env\Scripts\data"
output_dir=r"F:\vscode\stock_analysis\env\Scripts\yaml_to_csv"
os.makedirs(output_dir,exist_ok=True)

for month_folder in sorted(os.listdir(data_dir)):
    month_path=os.path.join(data_dir,month_folder)
    if not os.path.isdir(month_path):
        continue

    for yaml_file in sorted(os.listdir(month_path)):
        if yaml_file.endswith('.yaml') or yaml_file.endswith('.yml'):
            file_path=os.path.join(month_path,yaml_file)

            with open(file_path,'r') as f:
                daily_data=yaml.safe_load(f)
                
            date_str = Path(yaml_file).stem.split("_")[0]
            month_str = date_str[:7]

            csv_file_path = os.path.join(output_dir,f"{date_str}.csv")
            if isinstance(daily_data,list):
                df=pd.json_normalize(daily_data)
            else:
                print(f"Warning: The YAML file is not a list at {file_path}. check the format.")
                continue

            df['Date'] = date_str
            df['Month'] = month_str
            df.to_csv(csv_file_path,index=False)
            print(f"Converted {yaml_file} to CSV at {csv_file_path}")
print("All Yaml File converted to CSV successfully")
```
