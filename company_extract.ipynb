```python
import os
import pandas as pd

input_folder = r"F:\vscode\stock_analysis\env\Scripts\yaml_to_csv"
output_folder = r"F:\vscode\stock_analysis\env\Scripts\segregated_csv"
os.makedirs(output_folder,exist_ok=True)

symbol_data = {}

for filename in os.listdir(input_folder):
    if filename.endswith(".csv"):
        filepath = os.path.join(input_folder,filename)
        df= pd.read_csv(filepath)

        if 'Ticker' not in df.columns:
            print(f"Missing 'Ticker' column in {filename}, skip")
            continue
        
        for _, row in df.iterrows():
            symbol = row['Ticker']
            symbol_data.setdefault(symbol, []).append(row)

for symbol, rows in symbol_data.items():
    df = pd.DataFrame(rows)
    df.sort_values(by='Date', inplace=True)
    df.to_csv(os.path.join(output_folder, f"{symbol}.csv"),index=False)
print(f"All 50 company CSV files created in: \n{output_folder}")

```

```python
import os
import pandas as pd

input_folder = r"F:\vscode\stock_analysis\env\Scripts\segregated_csv"

for filename in os.listdir(input_folder):
    if filename.endswith(".csv"):
        filepath = os.path.join(input_folder,filename)

        df=pd.read_csv(filepath)

        if 'Date' in df.columns:
            df.drop(columns=['Date'], inplace=True)
            df.to_csv(filepath,index=False)
            print(f"'Date' column removed from {filename}")
        else:
            print(f"No 'Date' column in {filename}, skipped.")
print("All Files Processed")

```
